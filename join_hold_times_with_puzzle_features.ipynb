```python
import numpy as np
import pandas as pd
```

```python
member_info = pd.read_csv('data/member_clean.csv', index_col=0)
print(member_info.info())
print(member_info.dtypes)
member_info.head()
```

```python
puzzle_piece_info = pd.read_csv('data/pieces_by_puzzle_and_difficulty.csv', index_col = 0)
puzzle_piece_info['pack_name'] = puzzle_piece_info.name
puzzle_piece_info.drop('name', inplace=True, axis=1)
print(puzzle_piece_info.dtypes)
puzzle_piece_info.head()
```

```python
hold_times_joined = member_info.merge(puzzle_piece_info, left_on='pack_name', right_on='pack_name', how='left')
hold_times_joined
```
